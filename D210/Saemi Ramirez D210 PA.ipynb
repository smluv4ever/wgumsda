```python
# read the CSV file
df2 = pd.read_csv(r'C:\Users\Saemi\OneDrive\Education\WGU\D210 Representation and Reporting\PA\demographic.csv')
# check the data information
df2.info()
```

```python
df2 = df2.rename(columns={"RIDAGEYR":"Age", "RIAGENDR":"Gender", "DMDMARTL":"Marital"})
df2.info()
```

```python
dem_cols=["Age", "Gender", "Marital"]
df2[dem_cols]
```

```python
df2.Gender.replace({1:"Male", 2:"Female"}, inplace=True)
df2['Gender']
```

```python
df2.Marital.replace({1:"Married", 2:"Widowed", 3:"Divorced", 4:"Separated", 5:"Never Married", 6:"Living with Partner",
                         77:"Never Married", 99:"Never Married", ".":"Never Married"}, inplace=True)
df2.Marital.fillna("Never Married", inplace=True)
df2['Marital']
```

```python
df2[dem_cols]
```

```python
# read the CSV file
df3 = pd.read_csv(r'C:\Users\Saemi\OneDrive\Education\WGU\D210 Representation and Reporting\PA\questionnaire.csv')
# check the data information
df3.info()
```

```python
df3 = df3.rename(columns={"BPD035":"HighBlood", "MCQ025":"Asthma", "MCQ160F":"Stroke", "MCQ080E":"Overweight", 
                          "MCQ160A":"Arthritis", "DID040":"Diabetes"})
```

```python
q_cols = ["HighBlood", "Asthma", "Stroke", "Overweight", "Arthritis", "Diabetes"]
df3[q_cols]
```

```python
df3['HighBlood'] = df3['HighBlood'].notnull().map({True: 'Yes', False: 'No'})
df3['Asthma'] = df3['Asthma'].notnull().map({True: 'Yes', False: 'No'})
df3['Stroke'] = df3['Stroke'].notnull().map({True: 'Yes', False: 'No'})
df3['Overweight'] = df3['Overweight'].notnull().map({True: 'Yes', False: 'No'})
df3['Arthritis'] = df3['Arthritis'].notnull().map({True: 'Yes', False: 'No'})
df3['Diabetes'] = df3['Diabetes'].notnull().map({True: 'Yes', False: 'No'})
```

```python
df3[q_cols]
```

```python
ext_data = pd.concat([df2[dem_cols], df3[q_cols]], axis=1)
ext_data['Source'] = 'NHANES'
ext_data
```
