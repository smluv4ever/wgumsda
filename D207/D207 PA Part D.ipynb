```python
#import packages
import pandas as pd
from scipy import stats
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the medical csv file
df = pd.read_csv(r'C:\Users\Saemi\OneDrive\Education\WGU\D207 Exploratory Data Analysis (Python & R)\PA\medical_clean.csv')

#convert the categorical values to numeric
df['HighBlood_num'] = df['HighBlood']
HBN = {"HighBlood_num": {"No":0, "Yes":1}}
df.replace(HBN, inplace=True)

df['Overweight_num'] = df['Overweight']
ON = {"Overweight_num": {"No":0, "Yes":1}}
df.replace(ON, inplace=True)

#create the variables
VitD_levels = df['VitD_levels']
Initial_days = df['Initial_days']
HighBlood = df['HighBlood']
Overweight = df['Overweight']
HighBloodNum = df['HighBlood_num']
OverweightNum = df['Overweight_num']
```

```python
#Create correlation function for VitD_levels and Initial_days (Keith)
plt.scatter(VitD_levels, Initial_days)
plt.xlabel("Vitamin D Levels")
plt.ylabel("Days Hospitalized")
plt.show()
```

```python
pearson = stats.pearsonr(VitD_levels, Initial_days)
print(pearson)
```

```python
#Create t-test for VitD_levels and HighBlood
plt.bar(HighBlood, VitD_levels)
plt.xlabel("High Blood Pressure")
plt.ylabel("Vitamin D Levels")
plt.show()
```

```python
t_result1 = stats.ttest_ind(VitD_levels, HighBloodNum)
print(t_result1)
```

```python
#Createt-test for VitD_levels and Overweight
plt.bar(Overweight, VitD_levels)
plt.xlabel("Overweight")
plt.ylabel("Vitamin D Levels")
plt.show()
```

```python
t_result2 = stats.ttest_ind(VitD_levels, OverweightNum)
print(t_result2)
```

```python
#Create t-test for Initial_days and HighBlood
plt.bar(HighBlood, Initial_days)
plt.xlabel("High Blood Pressure")
plt.ylabel("Days Hospitalized")
plt.show()
```

```python
t_result3 = stats.ttest_ind(Initial_days, HighBloodNum)
print(t_result3)
```

```python
#Create t-test for Initial_days and Overweight
plt.bar(Overweight, Initial_days)
plt.xlabel("Overweight")
plt.ylabel("Days Hospitalized")
plt.show()
```

```python
t_result4 = stats.ttest_ind(Initial_days, OverweightNum)
print(t_result4)
```

```python
#Create chi-test for HighBlood and Overweight
sns.histplot(x=HighBlood, hue=Overweight, data=df)
```

```python
ratio = pd.crosstab(HighBlood, Overweight)
print(ratio)
```

```python
chi2 = chi2_contingency(ratio)
print(chi2)
```
